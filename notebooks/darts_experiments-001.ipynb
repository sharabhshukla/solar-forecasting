```python
import numpy as np
from darts import TimeSeries
from darts.models import LightGBMModel, CatBoostModel
from darts.utils.model_selection import train_test_split
```

```python
pv_data = context.catalog.load('preprocessed_raw_pv_data')
```

```python
pv_data.head()
```

```python
ts_data = TimeSeries.from_dataframe(pv_data, time_col='datetime', value_cols=pv_data.columns.drop('datetime'),
                                    fill_missing_dates=True, fillna_value=np.nan)
```

```python
ts_data['generation (w)'].plot()
```

```python
cbt_model = CatBoostModel(lags_past_covariates=24*1*12, lags=24*1*12, output_chunk_length=24*12, iterations=2000)
```

```python
train, test = train_test_split(ts_data, test_size=0.05)
```

```python
train_target = train['generation (w)']
test_target = test['generation (w)']
covariate_cols = train.columns.drop('generation (w)').tolist()
```

```python
cbt_model.fit(train_target, past_covariates=train[covariate_cols], verbose=True)
```

```python
cbt_model.predict(past_covariates=train[covariate_cols].append(test[covariate_cols]))
```
